**Problem Statement 1 : Smart Parking Lot Management System**

Design a function to manage a smart parking lot.

The system should:

Accept vehicle entry and exit logs


Calculate total parked vehicles


Identify peak parking usage


Alert if parking exceeds capacity


```python
def parking_mangement_system(parking_capacity, vehicle_log):
    total_parked_vehicle =0
    peak_usage =0

    for log in vehicle_log:
        if log == "IN":
            total_parked_vehicle += 1
        elif log == "OUT":
            if total_parked_vehicle > 0:
                total_parked_vehicle -=1

        if total_parked_vehicle > peak_usage:
            peak_usage = total_parked_vehicle

    print(f"Currently Parked Vehicle: {total_parked_vehicle}")
    print(f"Peak Parking Usage: {peak_usage}")

    if total_parked_vehicle > parking_capacity:
        print("Parking Status: Full (Over Capacity Alert!)")
    elif total_parked_vehicle == parking_capacity:
        print("Parking Status: Full")
    else:
        print("Parking Status: Available")

parking_capacity =50
vehicle_log =["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

parking_mangement_system(parking_capacity, vehicle_log)
```

**Problem Statement 2:Online Food Delivery Time Estimator**

Create a function that estimates delivery time based on:

Distance (km)


Weather condition


Traffic level


Apply delays dynamically and display final ETA.


```python
def FD_Estimator(distance, weather, traffic):
  est_Time=distance*5

  if weather == "Rain":
        est_Time +=5

  if traffic == "Medium":
        est_Time +=5
  elif traffic == "High":
        est_Time +=10
  print(f" Estimated Delivery Time : {est_Time} minutes")

FD_Estimator(8, "Rain", "High")
```

 **Problem Statement 3:Movie Theatre Seat Occupancy Analyzer**

Build a function that analyzes seat booking data and:

Calculates occupancy percentage


Determines if show is Housefull


Suggests opening additional shows


```python
def analyze_seat_occupancy(total_seats, booked_seats):

   number_of_booked_seats = len(booked_seats)

   occupancy_percentage = (number_of_booked_seats / total_seats) * 100

   print("Occupancy:", round(occupancy_percentage), "%")

   if occupancy_percentage ==100:
        print("Show Status: Housefull")
        print("Suggestion: Consider opening an additional show.")

   elif occupancy_percentage >=75:
        print("Show Status: Almost Full")

   else:
        print("Show Status: Seats Available")


total_seats =200
booked_seats =[1] * 150

analyze_seat_occupancy(total_seats, booked_seats)
```


 **Problem Statement 4:Cloud Server Load Classification System**

Create a function to classify server load based on CPU usage readings.

Rules:

Average CPU < 50% → Normal


50%–80% → Warning


80% → Critical





```python
def classify_server_load(cpu_readings):
    total_cpu =0

    for reading in cpu_readings:
        total_cpu +=reading

    average_cpu = total_cpu / len(cpu_readings)

    print("Average CPU Load:", round(average_cpu), "%")

    if average_cpu <50:
        print("Server Status: Normal")

    elif average_cpu >= 50 and average_cpu <= 80:
        print("Server Status: Warning")

    else:
        print("Server Status: Critical")

cpu_readings =[45, 60, 70, 85, 90]
classify_server_load(cpu_readings)
```


 **Problem Statement 5:Smart Classroom Resource Usage Monitor**

Design a function that tracks usage of classroom resources (projector, AC, lights) and identifies overuse patterns.


```python
def monitor_resource_usage(resource_usage):
    overused_resources =[]

    for resource, hours in resource_usage.items():
        if hours >8:
            overused_resources.append(resource)

    if overused_resources:
        print("Overused Resources:", ", ".join(overused_resources))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

resource_usage = {
    "Projector":6,
    "AC":9,
    "Lights":4
}

monitor_resource_usage(resource_usage)
```

 **Problem Statement 6: Online Event Registration Capacity Controller**

Create a function that manages event registrations by:

Tracking registrations


Preventing overbooking


Triggering waitlist mode


```python
def manage_event_registrations(event_capacity, total_registrations):
  confirmed =0
  waitlisted =0
  for i in range(1, total_registrations + 1):
        if confirmed < event_capacity:
            confirmed += 1
        else:
            waitlisted +=1

  print("Confirmed Registrations:", confirmed)
  print("Waitlisted Users:", waitlisted)

  if confirmed == event_capacity:
        print("Registration Status: Closed")
  else:
        print("Registration Status: Open")

event_capacity =100
registrations =105

manage_event_registrations(event_capacity, registrations)
```
