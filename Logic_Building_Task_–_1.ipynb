**1.User Login Check**

**Problem Statement:**

Given a username and password, check whether login is successful.

username = "admin"

password = "1234"


```python
# Predefined credentials
correct_username = "admin"
correct_password = "1234"

# User input
username = "admin"
password = "1234"

if username==correct_username and password==correct_password:
  print("Login Successful")
else:
  print("Invalid Credentials")
```

**2.Pass / Fail Analyzer**

**Problem Statement:**

Analyze student results.

marks = [45, 78, 90, 33, 60]


```python
marks = [45, 78, 90, 33, 60]
pass_count=0
fail_count=0

for mark in marks:
  if mark>=50:
    pass_count+=1
  else:
    fail_count+=1

print(f"The total number of student passed is : {pass_count}")
print(f"The total number of students failed is : {fail_count}")
```

**3.Simple Data Cleaner**

**Problem Statement:**

Clean and standardize user names.

names = [" Alice ", "bob", " CHARLIE "]



```python
names = [" Alice ", "bob", " CHARLIE "]
Names=[]

for name in names:
  Name=name.strip().lower()
  Names.append(Name)

print("Cleaned and standardized user names :")
print(Names)
```

**4.Message Length Analyzer**

**Problem Statement:**

Analyze message sizes.

messages = ["Hi", "Welcome to the platform", "OK"]


```python
messages = ["Hi", "Welcome to the platform", "OK"]

print("Message Length Analysis : \n ")
for message in messages:
  length=len(message)
  print(f" \tMessage : {message} | length :{length}  ")

  if length>10:
    print("\tMessage is longer than 10 characters ")
  print("\n")
```

**5.Error Message Detector**

**Problem Statement:**

Detect error messages from system logs.

logs = ["INFO", "ERROR", "WARNING", "ERROR"]


```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
error_count=0

for log in logs:
  if log=="ERROR":
    error_count+=1

print(f"The Total number of errors in system logs is : {error_count}")
```
