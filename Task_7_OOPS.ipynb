**Problem 1: Smart Light Controller**

Problem Statement

Create a class to control a smart light that can be turned ON or OFF and display its status.


```python
class Light_controller:
  def __init__(self,Light_name):
    self.Light_name=Light_name

  def turn_on(self):
        self.is_on = True
        print(f"{self.Light_name} is ON")

  def turn_off(self):
        self.is_on = False
        print(f"{self.Light_name} is OFF")

  def display_status(self):
        if self.is_on:
            print(f"{self.Light_name} is ON")
        else:
            print(f"{self.Light_name} is OFF")

light = Light_controller("Bedroom Light")
light.turn_on()
```

**Problem 2: Employee ID Card System**

 Problem Statement

Create a class that stores employee name, ID, and department and displays ID card details.


```python
class employee_card_system:
  def __init__(self,name,id,department):
    self.name=name
    self.id=id
    self.department=department

  def display_result(self):
     print("Employee ID card")
     print(f"Name : {self.name} \nID : {self.id} \nDepartment : {self.department}")


emp=employee_card_system('Rahul','EMP102','IT')
emp.display_result()
```

**Problem 3: Mobile Contact Record**

Problem Statement:

Create a class to store contact name and phone number and display contact information.


```python
class mobile_contact_record:
  def __init__(self,name,phone_number):
    self.name=name
    self.phone_number=phone_number

  def display_record(self):
    print("Contct Saved")
    print("Name : ",self.name)
    print("Phone :",self.phone_number)

customer=mobile_contact_record('Anita',9876543210)
customer.display_record()
```

**Problem 4: Product Price Tag Generator**

Problem Statement:

Create a class that stores product name and price and prints a formatted price tag.


```python
class price_tag_generator:
  def __init__(self,name,price):
    self.name=name
    self.price=price

  def formatted_result(self):
    print(f"Product : {self.name} ")
    print(f"Price : ₹{self.price}")

product1=price_tag_generator('Headphones',5999)
product1.formatted_result()
```

**Problem 5: Movie Rating Display System**

Problem Statement:

Create a class that stores movie name and rating and displays movie details.


```python
class movie_rating_system:
  def __init__(self,name,rating):
    self.name=name
    self.rating=rating

  def result(self):
    print(f"Movie : {self.name}")
    print(f"Rating : {self.rating}/5")

movie1=movie_rating_system('Inception',4.8)
movie1.result()
```

**Problem 6: Delivery Address Manager**

Problem Statement:

Create a class that stores customer name and delivery address and prints delivery details.
